**Obs: Para o Windows, baixe a versão .exe):**

### 2. Instalação no Linux

#### Caso já não esteja instalado, instalar o Java:

Acessar o endereço http://www.oracle.com/technetwork/java/javase/downloads/index.html para download da versão mais atual do Java.

#### Descompactar o arquivo neo4j-community-3.x.x-unix.tar.gz:

$ tar -xvzf neo4j-community-3.x.x-unix.tar.gz

#### Entrar no diretório "bin" da pasta descompactada:

$ cd neo4j-community-3.x.x/bin/

#### Iniciar o serviço do Neo4j:

$ ./neo4j start


#### Para interromper o serviço do Neo4j:

$ ./neo4j stop


### 2. Instalação no Windows

#### Caso já não esteja instalado, instalar o Java:

Acessar o endereço http://www.oracle.com/technetwork/java/javase/downloads/index.html para download da versão mais atualdo Java.

#### Executar o arquivo neo4j-community_windows_3_x_x.exe, para inicializar o instalador do Neo4j.

#### Siga as mensagens do processo de instalação, e no final, escolher a opção para executar o Neo4j.

Next...Next...Next...Finish.

#### Será aberto o aplicativo do Neo4j. Para iniciar o serviço do Neo4j, apertar o botão "Start".
